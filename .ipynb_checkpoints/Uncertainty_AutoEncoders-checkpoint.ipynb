##**Baseline 1 : Autoencoder**

```python
input_shape = X_train.shape[1]
```

```python
# Define the encoding layer
input_data = Input(shape=(input_shape,))
encoded = Dense(32, activation="relu")(input_data)
encoded = BatchNormalization()(encoded)
encoded = Dropout(0.5)(encoded)

# Decoding layer
decoded = Dense(input_shape, activation='sigmoid')(encoded)

autoencoder = Model(input_data, decoded)
```

```python
autoencoder.compile(optimizer='adam', loss='mean_squared_error')
```

```python
autoencoder.summary()
```

```python
early_stopping = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
autoencoder_train=autoencoder.fit(X_train, X_train,
                epochs=100,
                batch_size=256,
                shuffle=True,
                validation_data=(X_test, X_test),callbacks=[early_stopping],verbose=1)
```

```python
loss = autoencoder_train.history['loss']
val_loss = autoencoder_train.history['val_loss']
epochs = range(len(loss))
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
y_pred = autoencoder.predict(X_test)
```

```python
mse = np.mean(np.power(X_test - y_pred, 2), axis=1)
```

```python
plt.plot(mse, marker='o', linestyle='')
plt.xlabel('Sample Index')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('MSE Values for Reconstructed Samples')
plt.grid(True)
plt.show()
```

```python
y_predictions = y_pred[:, 0]
```

```python
threshold = np.percentile(mse, 85)

y_pred = (mse > threshold).astype(int)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
thresholds = np.linspace(min(mse), max(mse), num=100)
precision_list, recall_list, f1_list = [], [], []

# Calculate precision, recall, and F1-score for each threshold
for threshold in thresholds:
    predicted_anomalies = (mse > threshold).astype(int)
    precision = precision_score(y_test, predicted_anomalies)
    recall = recall_score(y_test, predicted_anomalies)
    f1 = f1_score(y_test, predicted_anomalies)

    precision_list.append(precision)
    recall_list.append(recall)
    f1_list.append(f1)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(thresholds, precision_list, label='Precision')
plt.plot(thresholds, recall_list, label='Recall')
plt.plot(thresholds, f1_list, label='F1-score')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.legend()
plt.title('Precision, Recall, and F1-score for different Thresholds')
plt.grid(True)
plt.show()
```

```python
def mc_dropout_predict(model, inputs, n_samples=100):
    """
    Perform Monte Carlo predictions with dropout enabled.
    
    :param model: The trained Keras model with Dropout layers
    :param inputs: Input data for making predictions
    :param n_samples: Number of Monte Carlo samples to draw
    :return: Numpy array of predictions
    """
    mc_predictions = [model(inputs, training=True).numpy() for _ in range(n_samples)]
    return np.array(mc_predictions)

# Perform Monte Carlo predictions
mc_predictions = mc_dropout_predict(autoencoder, df_test, n_samples=20)

# Compute mean and standard deviation (as the uncertainty measure)
mean_predictions = mc_predictions.mean(axis=0)
std_dev_predictions = mc_predictions.std(axis=0)

# Now you can print the mean prediction and associated uncertainty for a single data point
sample_index = 0  # Change this to the index of the sample you're interested in
mean_sample_prediction = mean_predictions[sample_index]
std_sample_prediction = std_dev_predictions[sample_index]

print(f"Mean prediction for sample {sample_index}: {mean_sample_prediction}")
print(f"Uncertainty (Standard Deviation) for sample {sample_index}: {std_sample_prediction}")

# Optionally, visualize the prediction and uncertainty as a bar plot
plt.figure(figsize=(14, 4))
plt.bar(range(len(mean_sample_prediction)), mean_sample_prediction, yerr=std_sample_prediction, alpha=0.5, capsize=5)
plt.title('Prediction with Uncertainty for Each Feature')
plt.xlabel('Feature Index')
plt.ylabel('Reconstruction Value')
plt.show()
```

# Deep Ensembles

```python

def create_autoencoder(input_shape):
    input_data = Input(shape=(input_shape,))
    encoded = Dense(32, activation="relu")(input_data)
    encoded = BatchNormalization()(encoded)
    encoded = Dropout(0.5)(encoded)
    decoded = Dense(input_shape, activation='sigmoid')(encoded)
    autoencoder = Model(input_data, decoded)
    autoencoder.compile(optimizer='adam', loss='mean_squared_error')
    return autoencoder

def train_ensemble(n_models, X_train, X_test, epochs, batch_size):
    models = []
    for _ in range(n_models):
        # Create a new model instance with random initialization
        model = create_autoencoder(df_train.shape[1])
        # Train the model
        early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
        model.fit(X_train, X_train, epochs=epochs, batch_size=batch_size,
                  shuffle=True, validation_data=(X_test, X_test),
                  callbacks=[early_stopping], verbose=1)
        models.append(model)
    return models

# Train the ensemble of autoencoders
n_models = 5  # Number of models in the ensemble
ensemble = train_ensemble(n_models, X_train, X_test, epochs=100, batch_size=256)

# Make predictions with each model in the ensemble
ensemble_predictions = [model.predict(X_test) for model in ensemble]

# Calculate mean and standard deviation across the ensemble predictions
mean_predictions = np.mean(ensemble_predictions, axis=0)
std_dev_predictions = np.std(ensemble_predictions, axis=0)

# Uncertainty estimation
uncertainty = std_dev_predictions.mean(axis=1)  # Mean uncertainty across all features
```
